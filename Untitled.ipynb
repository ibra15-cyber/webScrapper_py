```python
import requests
from bs4 import BeautifulSoup

request=requests.get("https://en.wikipedia.org/wiki/List_of_the_verified_oldest_people")
data = request.content
```

```python
soup=BeautifulSoup(data, "html.parser")

# print(soup.prettify)
```

```python
richest=soup.find("table", {"table","wikitable sortable"})
# print(richest)
# type(richest)
```

```python
richest=soup.find("table",{"table": "wikitable sortable"})   #does the same as above
```

```python
richest[0].find_all("h2")[0].text #finds the first table in richest.then all the h2 tags but returns the first
```

```python
for item in richest:
    print(item.find_all("p")[0].text) #iterates the main tag and for each main tage it returns the first p tag 
```

```python
import requests
from bs4 import BeautifulSoup

request = requests.get("https://www.investopedia.com/articles/investing/012715/5-richest-people-world.asp")
data=request.content
soup=BeautifulSoup(data, "html.parser")
soup.prettify
```

```python
richest=soup.find_all("ul", {'id':"mntl-toc__list_1-0"})
print(richest)
```

```python
richest=soup.find_all("div", {'class':"comp mntl-toc__list-item-heading mntl-block"})
print(richest)
```

```python
richest=soup.find_all("a", {'class':"mntl-toc__list-item-link mntl-text-link mntl-toc__list-item-link js-mntl-toc-link"})
print(list(richest))
```

```python
richest=soup.find_all("span", {'class':"link__wrapper"})[2]
print(richest)
```

```python
#this line is doing the same as above but using iteration 
for name in soup.find_all("ul", {'id':"mntl-toc__list_1-0"}):
    names=name.find_all("span")
    print(names)
```
